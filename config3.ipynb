```python
import random 
from numpy import zeros, sqrt, pi, exp,linspace, floor
import pylab 
from statistics import NormalDist
from scipy.optimize import curve_fit

L = 100
N = 1000
iter = 10000
Nlist = [n for n in range(N)]
stats = zeros([3*N, iter], dtype = int)
Ncum = []
NcumIt = [0]
Ncumc = [N]

history = zeros([3*N,iter], dtype = int)



for n in range(N):
    history[n,0], history[n+N,0], history[n+2*N,0] = random.randint(1,L-1), random.randint(1,L-1), random.randint(1,L-1)

```

```python
for iterations in range(1,iter):
    space = zeros([L+1,L+1, L+1], dtype = int)
    for j in Nlist:
        a = space[history[j,iterations-1], history[j+N, iterations-1], history[j+2*N, iterations-1]]
        if a>0:
            Nlist.remove(j)
            Ncum.append(j)
            for q in range(iterations, iter):
                history[j,q], history[j+N,q], history[j+2*N,q] = history[j,iterations-1], history[j+N,iterations-1], history[j+2*N, iterations-1]
            for i in Nlist:
                if history[j,iterations-1] == history[i,iterations-1] and history[j+N,iterations-1] == history[i+N,iterations-1] and history[j+2*N,iterations-1] == history[i+2*N,iterations-1] :
                    Nlist.remove(i)
                    for q in range(iterations, iter):
                        history[i,q], history[i+N,q],history[i+2*N,q] = history[i,iterations-1], history[i+N,iterations-1], history[i+2*N,iterations-1]
                    Ncum.append(i)
            

        if not a>0:
            space[history[j,iterations-1], history[j+N, iterations-1]]+=1
    NcumIt.append(iterations-1)
    Ncumc.append(len(Nlist))

    for n in Nlist:
        dir = random.randint(1,6)
        if dir == 1:
            history[n,iterations], history[n+N,iterations], history[n+2*N,iterations] = history[n,iterations-1]+1, history[n+N,iterations-1], history[n+2*N,iterations-1]
            if history[n,iterations]>L:
                history[n,iterations]-=2 

        if dir == 2:
            history[n,iterations], history[n+N,iterations], history[n+2*N,iterations] = history[n,iterations-1]-1, history[n+N,iterations-1], history[n+2*N,iterations-1]

            if history[n,iterations]<0:
                history[n,iterations]+=2 


        if dir == 3:
            history[n,iterations], history[n+N,iterations], history[n+2*N,iterations] = history[n,iterations-1], history[n+N,iterations-1]+1, history[n+2*N,iterations-1]

            if history[n+N,iterations]>L:
                history[n+N,iterations]-=2 

        if dir == 4:
            history[n,iterations], history[n+N,iterations], history[n+2*N,iterations] = history[n,iterations-1], history[n+N,iterations-1]-1, history[n+2*N,iterations-1]
            if history[n+N,iterations]<0:
                history[n+N,iterations]+=2 


        if dir == 5:
            history[n,iterations], history[n+N,iterations], history[n+2*N,iterations] = history[n,iterations-1], history[n+N,iterations-1], history[n+2*N,iterations-1]+1
            if history[n+2*N,iterations]>L:
                history[n+2*N,iterations]-=2 


        if dir == 6:
            history[n,iterations], history[n+N,iterations], history[n+2*N,iterations] = history[n,iterations-1], history[n+N,iterations-1], history[n+2*N,iterations-1]-1

            if history[n+2*N,iterations]<0:
                history[n+2*N,iterations]+=2 

    for n in range(N):
        stats[n,iterations], stats[n+N, iterations], stats[n+2*N, iterations] = int(history[n,iterations]-history[n,0]), int(history[n+N,iterations]-history[n+N,0]), int(history[n+2*N,iterations]-history[n+2*N,0])

```
